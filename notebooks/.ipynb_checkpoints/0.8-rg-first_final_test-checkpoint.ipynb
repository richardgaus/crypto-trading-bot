```python
%load_ext autoreload
%autoreload 2

import os
from pathlib import Path
import pandas as pd
import numpy as np

from definitions import REPO_ROOT
from src.models.strategies import RSIStoch200EMA, BuyAndHold
```

## 1. Evaluate strategy on all cryptocurrencies

Load cryptocurrency data

```python
DATA_DIR = Path("D:/binance_ohlcv/30min")
boundary_timestamp = pd.Timestamp('2021-01-01 00:00:00')  # Boundary between validation and test set
margin = pd.DateOffset(months=2)  # How long the validation set timeseries should be at least

validation_sets = {}
test_sets = {}
max_first_timestamp = boundary_timestamp - margin

files = os.listdir(DATA_DIR)
for file in files:
    df = pd.read_csv(DATA_DIR / file)
    df.index = pd.DatetimeIndex(df['open time'])
    if df.index[0] > max_first_timestamp:  # Skip assets without long enough history
        continue
    validation_sets[file[:-4]] = df[:boundary_timestamp][:-1]
    test_sets[file[:-4]] = df[boundary_timestamp:]
```

```python
len(validation_sets)
```

```python
pnl_history_dir = REPO_ROOT / 'reports'
(pnl_history_dir / 'pnl_history_rsi').mkdir(parents=True, exist_ok=True)
(pnl_history_dir / 'pnl_history_bnh').mkdir(parents=True, exist_ok=True)

strategy_rsi = RSIStoch200EMA(
    max_number_open_trades=1,
    min_period_to_last_low=2,
    max_period_to_last_low=100,
    min_absolute_slope_rsi=0.16,
    swing_low_margin=2,
    swing_low_factor=0.999,
    take_profit_factor=1.50
)
strategy_bnh = BuyAndHold()
strategy_results_rsi = {}
strategy_results_bnh = {}

for i, (asset, df) in enumerate(validation_sets.items()):
    print(f'{i+1:03}/{len(validation_sets)} {asset} ', end='')
#    if f'{asset}.csv' in os.listdir(pnl_history_dir / 'pnl_history_rsi'):
#        strategy_results_rsi[asset] = pd.Series.from_csv(
#            pnl_history_dir / 'pnl_history_rsi' / f'{asset}.csv'
#        )
#    else:
    strategy_results_rsi[asset] = strategy_rsi.apply(
        ohlcv_timeseries=df,
        asset_name=asset
    )
    strategy_results_rsi[asset].evaluation()
    strategy_results_rsi[asset].eval_stats['pnl_history'].to_csv(
        pnl_history_dir / 'pnl_history_rsi' / f'{asset}.csv'
    )
#    if f'{asset}.csv' in os.listdir(pnl_history_dir / 'pnl_history_bnh'):
#        strategy_results_bnh[asset] = pd.Series.from_csv(
#            pnl_history_dir / 'pnl_history_bnh' / f'{asset}.csv'
#        )
#    else:
    strategy_results_bnh[asset] = strategy_bnh.apply(
        ohlcv_timeseries=df,
        asset_name=asset
    )
    strategy_results_bnh[asset].evaluation()
    strategy_results_bnh[asset].eval_stats['pnl_history'].to_csv(
        pnl_history_dir / 'pnl_history_bnh' / f'{asset}.csv'
    )
    print(f'PNL sri: {strategy_results_rsi[asset].pnl:.3f}, PNL bnh: {strategy_results_bnh[asset].pnl:.3f}')
```

```python
pnl_rsi, pnl_bnh = [], []
for key, item in strategy_results_rsi.items():
    pnl_rsi.append(item.pnl)
for key, item in strategy_results_bnh.items():
    pnl_bnh.append(item.pnl)
```
